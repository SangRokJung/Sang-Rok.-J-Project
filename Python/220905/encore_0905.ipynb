```python
jeju_p = jeju_1[7:].copy()
print(jeju_p[:5])
jeju_t = jeju_2[7:].copy()
print(jeju_t[:5])
jeju_h = jeju_3[16:].copy()
print(jeju_h[:5])
```

```python
# 날짜,평균기온(℃), 최저기온(℃), 최고기온(℃), 강수량(mm), 평균습도(%rh)로 jeju에 저장
jeju = []
for t, p, h in zip(jeju_t, jeju_p, jeju_h):
    t = t.split(',')
    p = p.split(',')
    h = h.split(',')
    jeju.append([t[0], t[2], t[3], t[4], p[2], h[3], h[4]])
jeju[:5]
```

```python
# 날짜,평균기온(℃), 최저기온(℃), 최고기온(℃), 강수량(mm), 평균습도(%rh)로 jeju에 저장
jeju = []
for t, p, h in zip(jeju_t, jeju_p, jeju_h):
    t = t.split(',')
    p = p.split(',')
    h = h.split(',')
    jeju.append([t[0], t[2], t[3], t[4][:-1], p[2][:-1], h[3], h[4][:-1]])
jeju[:5]
```

```python
# 2021 제주 최저 기온 및 평균 최저 기온 확인하기
low_temp = []
for day in jeju[1:]:
    low_temp.append(float(day[2]))
print('2021 제주 최저기온: {:5.2f}℃'.format(min(low_temp)))
print('2021 제주 평균기온: {:5.2f}℃'.format(sum(low_temp) / len(low_temp)))
```

```python
# 2021 제주 월별 평균 최저 기온 확인하기
month_temp_low = {}
for month in range(1,13):
    m = []
    for day in jeju[1:]:
        if month == int(day[0][5:7]):
            m.append(float(day[2]))
    temp_avg = '%.2f℃' %(sum(m)/len(m))
    month_temp_low[f'{month}월'] = temp_avg
print(month_temp_low)
```

```python
# 2021 제주 월별 평균 최고 기온 및 최고 기온 월 확인하기
month_temp_high = []
for month in range(1,13):
    m = []
    for day in jeju[1:]:
        if month == int(day[0][5:7]):
            m.append(float(day[3]))
    temp_avg = round(sum(m)/len(m), 2)
    month_temp_high.append(temp_avg)

for no, i in enumerate(month_temp_high):
    print(f'{no+1}월: {i}℃')
# 가장 높은 평균 최고기온을 기록한 월의 평균 온도?
hot = max(month_temp_high)
print(f'최대 기온 달: {month_temp_high.index(hot)+1}월, {hot}℃')
```

```python
# 가장 비가 많이 오는 월?
month_rain_dict = {}
month_rain_list = []
for month in range(1,13):
    m = []
    for i in jeju[1:]:
        if month == int(i[0][5:7]):
            if i[4] == '':
                i[4] = 0
            m.append(float(i[4]))
    rain_avg = round(sum(m)/len(m), 2)
    month_rain_dict[f'{month}월'] = rain_avg
    month_rain_list.append(rain_avg)
print('월별 강수량:', month_rain_dict)
hot = max(month_rain_list)
print(f'최대 강수량: {month_rain_list.index(hot)+1}월, {hot}mm')
```

```python
# 가장 습도가 낮은, 높은 월?
month_humidity = {}
for month in range(1,13):
    m = []
    for i in jeju[1:]:
        if month == int(i[0][5:7]):
            if i[5] == '':
                i[5] = 0
            m.append(float(i[5]))
    humidity_avg = round(sum(m)/len(m), 2)
    month_humidity[f'{month}월'] = humidity_avg
print('월별 습도:', month_humidity)
print('최저 습도:', min(month_humidity.values()))
print('최고 습도:', max(month_humidity.values()))
```

```python
class Circle:
    pi = 3.141592
    def __init__(self, radius):
        self.radius = radius
    def getArea(self):
        return Circle.pi * (self.radius ** 2)
    def getCircum(self):
        return 2 * Circle.pi * self.radius
```

```python
class EngSentence:
    def __init__(self, sentence):
        self.sentence = sentence
        self.length = len(sentence)
    def reverse(self):
        tmp = str()
        for i in range(1, self.length+1):
            tmp += self.sentence[-i]
        return tmp
            
    def insertHypen(self):
        tmp = str()
        for i in self.sentence:
            if i == ' ':
                tmp += '-'
            else:
                tmp += i
        return tmp
```

### 메소드 오버라이딩 예제 Person

##### 1. 초기값으로 name, age를 인스턴스 속성을 가진 Person 클래스를 생성 후 확인합니다

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def printInfo(self):
        print('name:', self.name)
        print('age :', self.age)
```

```python
p1 = Person('홍길동', 99)
p2 = Person('토끼', 200)
```

```python
# print(p1.name, p1.age)
# print(p2.name, p2.age)
p1.printInfo()
p2.printInfo()
```

```python
class Student(Person):
    def __init__(self, name, age, school, major):
        super().__init__(name, age)
        self.school = school
        self.major = major
    def printInfo(self):
        super().printInfo()
        print('school:', self.school)
        print('major:', self.major)
    def __str__(self):
        return f'{self.school}에 {self.major} 전공인 나이 {self.age}의 {self.name}입니다.'
```

```python
s1 = Student('홍길동', 99, 'encore', 'bigdata')
s2 = Student('토끼', 200, 'encore', 'bigdata')
```

```python
s1.printInfo()
s2.printInfo()
```

```python
# 다중 상속
class Father:
    def __init__(self):
        print('Father 생성자')
    def getFather(self, father):
        self.father = father
        return self.father

class Mother:
    def __init__(self):
        print('Mother 생성자')
    def getMother(self, mother):
        self.mother = mother
        return self.mother

class Child:
    def __init__(self):
        print('Child 생성자')
    def getChild(self, child):
        self.child = child
        return self.child
```

```python
f1 = Father()
m1 = Mother()
c1 = Child()
```

```python
print(f1.getFather('아버지'))
print(m1.getMother('어머니'))
print(c1.getChild('자식'))
```

```python
class Child(Father, Mother):
    def __init__(self):
        super().__init__()
        print('Child 생성자')
    def getChild(self, child):
        self.child = child
        return self.child
```
